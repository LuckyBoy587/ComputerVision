```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def loss_function(w):
    return w ** 2

def gradient(w):
    return 2 * w

def run_gradient_descent(w_start, alpha, steps):
    w_history = [w_start]
    w = w_start
    for _ in range(steps):
        w = w - alpha * gradient(w)
        w_history.append(w)
    return np.array(w_history)
```

```python
w_start = 10.0
steps = 20
learning_rates = [0.01, 0.1, 1.1]

results = {}
for alpha in learning_rates:
    results[alpha] = run_gradient_descent(w_start, alpha, steps)

for alpha in learning_rates:
    w_final = results[alpha][-1]
    loss_final = loss_function(w_final)
    print(f"alpha = {alpha} -> After {steps} steps: w ≈ {w_final:.4f}, Loss ≈ {loss_final:.4f}")
```

```python
plt.figure(figsize=(10, 6))
iterations = np.arange(steps + 1)

plt.plot(iterations, results[0.01], label="alpha = 0.01 (too slow)", color="blue", marker="o")
plt.plot(iterations, results[0.1], label="alpha = 0.1 (appropriate)", color="green", marker="s")
plt.plot(iterations, results[1.1], label="alpha = 1.1 (too large / diverging)", color="red", marker="x")

plt.title("Gradient Descent: Weight Convergence vs. Iterations")
plt.xlabel("Iteration")
plt.ylabel("Weight (w)")
plt.grid(True)
plt.legend()
plt.show()
```

### Analysis

1. **What happens at $\alpha = 1.1$? Why does this occur?**
   - At $\alpha = 1.1$, the weight $w$ alternates sign at each step and grows exponentially in magnitude (diverges).
   - This occurs because the update formula is $w_{t+1} = w_t(1 - 2\alpha)$. With $\alpha = 1.1$, the multiplier is $1 - 2(1.1) = -1.2$. Since the absolute value $|-1.2| = 1.2 > 1$, the magnitude $|w|$ grows by $20\%$ at each step, and the negative sign causes the value to alternate between positive and negative.

2. **What value of $\alpha$ would cause the weight to oscillate but not diverge?**
   - For the weight to oscillate, the update factor $1 - 2\alpha$ must be negative, requiring $\alpha > 0.5$.
   - To not diverge, the magnitude of the update factor $|1 - 2\alpha|$ must be less than or equal to $1$, requiring $\alpha \le 1.0$.
   - Specifically, $\alpha = 1.0$ causes the weight to oscillate between $10.0$ and $-10.0$ indefinitely with a constant magnitude (neither converging nor diverging).
