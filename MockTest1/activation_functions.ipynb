```python
import numpy as np

Z = [-2.0, -0.5, 0.0, 1.0, 3.5]
def sigmoid(val):
    return 1 / (1 + np.exp(-val))

def relu(val):
    return max(0, val)
```

```python
for z in Z:
    print(f"Z = {z} -> sigmoid = {sigmoid(z):.3f} | relu = {relu(z)}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(range(-5, 5), [sigmoid(z) for z in range(-5, 5)], label="Sigmoid")
plt.plot(range(-5, 5), [relu(z) for z in range(-5, 5)], label="Relu")
plt.legend()
```
