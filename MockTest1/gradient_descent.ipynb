```python
w = 0.0
dl_dw = lambda val: 2 * (val - 3)
alpha = 0.1

loss = lambda val: (val - 3) ** 2
loss_values = []
for i in range(11):
    print(f"Iteration {i}: w = {w:.6f}, loss = {loss(w):.6f}")
    loss_values.append(loss(w))
    w = w - alpha * dl_dw(w)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.plot(loss_values)
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.grid(True)
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))

for i, v in enumerate(loss_values):
    plt.annotate(f"{v:.2f}", (i, v), textcoords="offset points", xytext=(0, 8), ha='center')
```
