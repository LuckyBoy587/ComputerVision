# Task 1: Perceptron Concept & Components

### Definition of Perceptron
A Perceptron is the simplest form of an artificial neural network used for binary classification. It models a single biological neuron by taking multiple inputs, applying weights, adding a bias, and passing the result through an activation function.

### Components
- **Inputs**: Features representing the input data.
- **Weights**: Coefficients representing the strength of connection.
- **Bias**: An offset that shifts the activation function.
- **Activation Function**: Introduces non-linearity to determine output.

### Mathematical Representation
$$Z = (w_1 \cdot x_1 + w_2 \cdot x_2) + b$$
$$A = f(Z)$$

```
x1 = 1.0
x2 = 2.0
w1 = 0.5
w2 = -0.6
b = 0.1

z = (w1 * x1 + w2 * x2) + b
print(f"Z = {z}")
```

```
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

x_val = np.array([-5, -2, -1, 0, 1, 2, 5])
print("Sigmoid outputs:", sigmoid(x_val))
print("ReLU outputs:", relu(x_val))
```

```
plt.plot(x_val, sigmoid(x_val), label="Sigmoid")
plt.plot(x_val, relu(x_val), label="ReLU")
plt.title("Sigmoid vs ReLU")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

### Activation Functions Comparison

| Feature | Sigmoid | ReLU |
|---|---|---|
| **Output Range** | (0, 1) | [0, $\infty$) |
| **Behavior** | S-shaped curve that maps inputs to probability-like values, but suffers from vanishing gradient problem for very large/small values. | Linear for positive values and zero for negative values, resolving vanishing gradient but subject to dying ReLU. |
| **Use Cases** | Binary classification output layer. | Hidden layers of deep neural networks. |

```
X = np.array([1.0, 2.0])
W1 = np.array([[0.5, -0.6], [0.8, 0.2]])
b1 = np.array([0.1, -0.2])
W2 = np.array([0.7, -1.2])
b2 = 0.3

z1_1 = np.dot(W1[0], X) + b1[0]
z1_2 = np.dot(W1[1], X) + b1[1]

a1_1 = relu(z1_1)
a1_2 = relu(z1_2)

print(f"Z1 = {z1_1:.1f}, A1 = {a1_1:.1f}")
print(f"Z2 = {z1_2:.1f}, A2 = {a1_2:.1f}")

a1 = np.array([a1_1, a1_2])
z_out = np.dot(W2, a1) + b2
final_output = sigmoid(z_out)

print(f"Z_out = {z_out:.1f}")
print(f"Final Output = {final_output:.5f}")
```

```
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(2, activation="relu", input_shape=(2,)),
    Dense(1, activation="sigmoid")
])

w1_tf = np.array([[0.5, 0.8], [-0.6, 0.2]])
b1_tf = np.array([0.1, -0.2])
w2_tf = np.array([[0.7], [-1.2]])
b2_tf = np.array([0.3])

model.layers[0].set_weights([w1_tf, b1_tf])
model.layers[1].set_weights([w2_tf, b2_tf])

model.summary()

x_tf = np.array([[1.0, 2.0]])
prediction = model.predict(x_tf)
print(f"Prediction: {prediction[0][0]:.5f}")
```
