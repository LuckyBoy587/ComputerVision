# Task 1: Loss Functions Theory & Implementation

### Mean Squared Error (MSE)
MSE measures the average of the squares of the errors—that is, the average squared difference between the estimated values and the actual value. It is commonly used for regression tasks.
$$\text{MSE} = \frac{1}{N} \sum_{i=1}^N (y_i - \hat{y}_i)^2$$

### Binary Cross Entropy (BCE)
BCE measures the performance of a classification model whose output is a probability value between 0 and 1. It penalizes predictions that are confident but incorrect.
$$\text{BCE} = -\frac{1}{N} \sum_{i=1}^N \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]$$

```
import numpy as np
import matplotlib.pyplot as plt

y_pred = np.array([0.2, 0.7, 0.9])
y_true = np.array([0, 1, 1])

mse_val = np.mean((y_true - y_pred) ** 2)
bce_val = -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

print(f"MSE: {mse_val:.5f}")
print(f"BCE: {bce_val:.5f}")
```

```
x_init = 5.0
iterations = 20

lr1 = 0.1
lr2 = 0.01

x_lr1 = []
x = x_init
for _ in range(iterations):
    grad = 2 * x
    x = x - lr1 * grad
    x_lr1.append(x)

x_lr2 = []
x = x_init
for _ in range(iterations):
    grad = 2 * x
    x = x - lr2 * grad
    x_lr2.append(x)

print("LR = 0.1 values:", [round(val, 4) for val in x_lr1])
print("LR = 0.01 values:", [round(val, 4) for val in x_lr2])
```

```
plt.plot(x_lr1, label="LR=0.1")
plt.plot(x_lr2, label="LR=0.01")
plt.title("Gradient Descent Optimization")
plt.xlabel("Iterations")
plt.ylabel("x value")
plt.grid(True)
plt.legend()
plt.show()
```

### Observation of Learning Rate Effect
- **Higher Learning Rate ($\alpha = 0.1$)**: Leads to rapid convergence towards the minimum ($x = 0$) within fewer iterations.
- **Lower Learning Rate ($\alpha = 0.01$)**: Converges much slower, requiring significantly more iterations to reach the minimum.

```
x_val = 2.0
y_true = 4.0
w = 1.0
lr = 0.1

y_pred = w * x_val
loss = (y_pred - y_true) ** 2
dw = 2 * (y_pred - y_true) * x_val
w_new = w - lr * dw

print(f"Prediction: {y_pred:.1f}")
print(f"Loss: {loss:.1f}")
print(f"Gradient: {dw:.1f}")
print(f"Updated Weight: {w_new:.1f}")
```

### Task 4: Weight Update & Learning Rate Effect
The updated weight is calculated as:
$$w_{\text{new}} = w_{\text{old}} - \eta \cdot dw = 1.0 - 0.1 \cdot (-8) = 1.8$$

### Learning Rate Effect
- If the learning rate is **too small**, the weight updates are tiny, leading to extremely slow convergence.
- If the learning rate is **too large**, the updates can overshoot the minimum, causing the loss to oscillate or even diverge.

```
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

X = np.array([[1.0], [2.0], [3.0]])
y = np.array([[2.0], [4.0], [6.0]])

model = Sequential([
    Dense(1, input_shape=(1,))
])

model.layers[0].set_weights([np.array([[0.84]]), np.array([0.0])])

model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01), loss="mean_squared_error")
history = model.fit(X, y, epochs=50, verbose=1)
```

```
plt.plot(history.history["loss"])
plt.title("Loss vs Iterations")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
```

### Comparison of Learning Rates in Training
- **Learning Rate = 0.1**: Can lead to unstable training or divergence because the step size is too large for the loss landscape.
- **Learning Rate = 0.01**: Optimal learning rate that results in smooth, stable convergence to zero loss within 50 epochs.
- **Learning Rate = 0.001**: Too conservative, resulting in extremely slow learning where the loss remains high after 50 epochs.
