```
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
```

```
np.random.seed(42)
tf.random.set_seed(42)

X = np.random.rand(100, 1)
noise = np.random.normal(0, 0.1, (100, 1))
y = 2 * X + noise

split_idx = int(0.8 * len(X))
X_train, X_val = X[:split_idx], X[split_idx:]
y_train, y_val = y[:split_idx], y[split_idx:]

plt.scatter(X, y)
plt.title("Synthetic Dataset")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```
model_task1 = Sequential([
    Dense(8, activation="relu", input_shape=(1,)),
    Dense(8, activation="relu"),
    Dense(1)
])
model_task1.compile(optimizer="sgd", loss="mse")
history_task1 = model_task1.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=50, verbose=0)
model_task1.summary()
```

### Layer Stacking & How Depth Affects Learning
Layer stacking involves adding multiple layers sequentially to help the model learn complex, non-linear relationships.
As depth increases, the network can capture higher-level abstractions. However, excessive depth can lead to vanishing gradients and overfitting.

```
model_task2 = Sequential([
    Dense(128, activation="relu", input_shape=(1,)),
    Dense(128, activation="relu"),
    Dense(64, activation="relu"),
    Dense(1)
])
model_task2.compile(optimizer="adam", loss="mse")
history_task2 = model_task2.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=100, verbose=0)
```

### Overfitting vs Underfitting
Overfitting occurs when a model learns the training data too well, including the noise, resulting in low training error but high validation error.
Underfitting occurs when the model is too simple to capture the underlying pattern of the data, leading to high error on both training and validation sets.

```
model_task3 = Sequential([
    Dense(128, activation="relu", input_shape=(1,)),
    Dropout(0.3),
    Dense(128, activation="relu"),
    Dropout(0.3),
    Dense(64, activation="relu"),
    Dropout(0.3),
    Dense(1)
])
model_task3.compile(optimizer="adam", loss="mse")
history_task3 = model_task3.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=100, verbose=0)
```

### How Dropout Prevents Overfitting
Dropout randomly deactivates a fraction of neurons during training (with probability p). This forces the network to learn redundant representations and prevents co-adaptation of features, acting as a form of regularization.

```
model_task4 = Sequential([
    Dense(128, input_shape=(1,)),
    BatchNormalization(),
    Dense(128, activation="relu"),
    BatchNormalization(),
    Dense(64, activation="relu"),
    BatchNormalization(),
    Dense(1)
])
model_task4.compile(optimizer="adam", loss="mse")
history_task4 = model_task4.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=50, verbose=0)
```

### How Batch Normalization Works
Batch Normalization standardizes the inputs to each layer for each mini-batch. This stabilizes the learning process, reduces internal covariate shift, allows for higher learning rates, and accelerates convergence.

```
model_sgd = Sequential([
    Dense(8, activation="relu", input_shape=(1,)),
    Dense(8, activation="relu"),
    Dense(1)
])
model_sgd.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01), loss="mse")
history_sgd = model_sgd.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=50, verbose=0)

model_adam = Sequential([
    Dense(8, activation="relu", input_shape=(1,)),
    Dense(8, activation="relu"),
    Dense(1)
])
model_adam.compile(optimizer="adam", loss="mse")
history_adam = model_adam.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=50, verbose=0)
```

```
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history_task2.history["val_loss"], label="Overfit")
plt.plot(history_task3.history["val_loss"], label="Dropout")
plt.plot(history_task4.history["val_loss"], label="BatchNorm")
plt.title("Model Comparison")
plt.xlabel("Epochs")
plt.ylabel("Validation Loss")
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history_sgd.history["loss"], label="SGD")
plt.plot(history_adam.history["loss"], label="Adam")
plt.title("Optimizer Comparison")
plt.xlabel("Epochs")
plt.ylabel("Training Loss")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
