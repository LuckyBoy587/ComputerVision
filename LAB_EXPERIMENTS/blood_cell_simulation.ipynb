```python
import zipfile
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
zip_path = 'blood-cells.zip'
image_to_extract = None

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    for file_info in zip_ref.infolist():
        if file_info.filename.lower().endswith(('.png', '.jpg', '.jpeg')) and not file_info.is_dir():
            image_to_extract = file_info.filename
            zip_ref.extract(image_to_extract, path='.')
            break

print("Extracted image:", image_to_extract)
```

```python
image = cv2.imread(image_to_extract)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(thresh, cmap='gray')
plt.title("Simulated Prediction")
plt.axis('off')

plt.tight_layout()
plt.show()
```
