# Task 1: Computation

Given bounding box coordinates:
- $x_{min} = 10$
- $y_{min} = 20$
- $x_{max} = 50$
- $y_{max} = 80$

Computation:
- $\text{Width} = x_{max} - x_{min} = 50 - 10 = 40$
- $\text{Height} = y_{max} - y_{min} = 80 - 20 = 60$

```python
%%bash
if [ -d "./Covid19-dataset" ]; then
    echo "Dataset already exists, skipping download."
else
    echo "Dataset not found, downloading..."
    kaggle datasets download -d pranavraikokte/covid19-image-dataset --force
    unzip -q covid19-image-dataset.zip -d .
fi
```

```python
import os
import glob
import cv2
import matplotlib.pyplot as plt
```

```python
dataset_path = 'Covid19-dataset'
image_paths = glob.glob(os.path.join(dataset_path, 'train', '*', '*'))[:5]

x_min, y_min = 10, 20
x_max, y_max = 50, 80

plt.figure(figsize=(8, 20))

for i, img_path in enumerate(image_paths):
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    cv2.rectangle(img_rgb, (x_min, y_min), (x_max, y_max), (255, 0, 0), 2)

    plt.subplot(5, 1, i + 1)
    plt.imshow(img_rgb)
    plt.title(f"Image {i + 1}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
