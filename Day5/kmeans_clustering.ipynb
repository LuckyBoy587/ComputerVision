```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
data = load_breast_cancer(as_frame=True)
df = data.frame

df_features = df.drop(columns=["target"])
df_cleaned = df_features.fillna(df_features.mean())

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_cleaned)
```

```python
wcss = []
silhouette_scores = []
k_range = list(range(2, 11))

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(scaled_features, kmeans.labels_))

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(k_range, wcss, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method')

plt.subplot(1, 2, 2)
plt.plot(k_range, silhouette_scores, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores')

plt.tight_layout()
plt.show()
```

```python
best_k = k_range[np.argmax(silhouette_scores)]
kmeans_best = KMeans(n_clusters=best_k, random_state=42, n_init=10)
kmeans_best.fit(scaled_features)
```

```python
print(f"Best k determined by Silhouette Score: {best_k}")
print("Comparing Elbow vs Silhouette results:")
for k, w, s in zip(k_range, wcss, silhouette_scores):
    print(f"k = {k} | WCSS = {w:.2f} | Silhouette Score = {s:.4f}")

print("\nInterpretation:")
print("The Elbow method shows that WCSS decreases as k increases, with a notable elbow or bending point around k=2.")
print("The Silhouette method reaches its maximum score at k=2, indicating the highest cluster cohesion and separation.")
print("Combining both, k=2 is the optimal number of clusters, which aligns perfectly with the true target classes (malignant and benign).")
```
