```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding, Dense, Softmax
```

```python
reviews = [
    "The application provides stable performance during daily operations",
    "There are multiple issues affecting the overall system reliability",
    "User interface is intuitive and easy to understand",
    "The response time is slower than expected under peak load",
    "System updates improved performance and fixed previous issues",
    "The documentation is unclear and lacks sufficient details",
    "The platform maintains consistent functionality across sessions",
    "There are occasional failures during transaction processing",
    "The design is structured and follows standard guidelines",
    "Performance degradation is observed during extended usage"
]

selected_review = reviews[0]
review_words = selected_review.split()[:100]
processed_review = " ".join(review_words)

tokenizer = Tokenizer()
tokenizer.fit_on_texts([processed_review])

sequence = tokenizer.texts_to_sequences([processed_review])[0]
padded_sequence = pad_sequences([sequence], maxlen=100, padding='post')

print("Selected Review:", processed_review)
print("Sequence Representation:", sequence)
print("Padded Sequence Shape:", padded_sequence.shape)
```

```python
tf.random.set_seed(42)

vocab_size = len(tokenizer.word_index) + 1
embedding_dim = 8

embedding_layer = Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=100)
dense_layer = Dense(1)
softmax_layer = Softmax(axis=1)

input_tensor = tf.constant(padded_sequence, dtype=tf.int32)
embedded_seq = embedding_layer(input_tensor)
attention_scores = dense_layer(embedded_seq)
attention_weights = softmax_layer(attention_scores)

print("Attention Scores Shape:", attention_scores.shape)
print("Attention Weights Shape:", attention_weights.shape)
```

```python
tokenized_words = [tokenizer.index_word[idx] for idx in sequence]

print(f"{'Word':<15} | {'Attention Weight':<20}")
print("-" * 38)
for idx, word in enumerate(tokenized_words):
    weight = attention_weights[0, idx, 0].numpy()
    print(f"{word:<15} | {weight:.8f}")
```
