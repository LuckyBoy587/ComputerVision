```python
import zipfile
import os
import cv2
import matplotlib.pyplot as plt
```

```python
!kaggle datasets download -d paultimothymooney/blood-cells
```

```python
zip_path = 'blood-cells.zip'
extract_dir = 'blood_cells_extracted'

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    image_files = [f for f in zip_ref.namelist() if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    target_img = image_files[0]
    zip_ref.extract(target_img, path=extract_dir)
    img_path = os.path.join(extract_dir, target_img)
```

```python
image = cv2.imread(img_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, thresholded = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image_rgb)
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(thresholded, cmap="gray")
axes[1].set_title("Simulated Prediction")
axes[1].axis("off")

plt.tight_layout()
plt.show()
```
