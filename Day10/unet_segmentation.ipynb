### Task 1: Computation

For an input image of size 128 x 128:
- After MaxPooling (2x2): Output size is 64 x 64
- After UpSampling (2x2): Output size is 128 x 128

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
image_path = 'ComputerVision/data/brain_tumor_dataset/yes/Y1.jpg'
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
resized = cv2.resize(gray, (128, 128))
normalized = resized / 255.0
input_tensor = np.expand_dims(np.expand_dims(normalized, axis=0), axis=-1)
```

```python
inputs = layers.Input(shape=(128, 128, 1))
c1 = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(inputs)
p1 = layers.MaxPooling2D((2, 2))(c1)
c2 = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(p1)
u1 = layers.UpSampling2D((2, 2))(c2)
c3 = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(u1)
outputs = layers.Conv2D(1, (1, 1), activation='sigmoid', padding='same')(c3)
model = models.Model(inputs=[inputs], outputs=[outputs])
model.summary()
```

```python
prediction = model.predict(input_tensor)
predicted_mask = np.squeeze(prediction)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(resized, cmap='gray')
plt.title('Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(predicted_mask, cmap='gray')
plt.title('Predicted Mask')
plt.axis('off')
plt.show()
```
