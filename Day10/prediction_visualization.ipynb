# Prediction Analysis and Visualization

### Task 1: Computation

Given:
- Prediction values = [0.2, 0.8, 0.6, 0.1, 0.9]
- Threshold = 0.5

Results:
- Binary outputs: [0, 1, 1, 0, 1]
- Number of positive predictions: 3

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
predictions = np.array([0.2, 0.8, 0.6, 0.1, 0.9])
threshold = 0.5
binary_outputs = (predictions >= threshold).astype(int)
positive_count = np.sum(binary_outputs)

print("Predictions:", predictions)
print("Binary outputs:", binary_outputs.tolist())
print("Positive count:", positive_count)
```

```python
plt.figure(figsize=(8, 5))
for i, pred in enumerate(predictions):
    plt.scatter(i, pred, s=100)
plt.axhline(y=threshold, color="tab:blue", linestyle="-")
plt.title("Prediction Visualization")
plt.xlabel("Index")
plt.ylabel("Prediction Value")
plt.show()
```
