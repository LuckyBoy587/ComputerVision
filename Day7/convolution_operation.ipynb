```python
import numpy as np
from tensorflow.keras.datasets import mnist
```

```python
(mnist_x, _), _ = mnist.load_data()
```

```python
image = mnist_x[0]
h, w = image.shape
patch = None
for r in range(h - 2):
    for c in range(w - 2):
        p = image[r:r+3, c:c+3]
        if np.count_nonzero(p) >= 5 and np.max(p) > 200:
            patch = p
            break
    if patch is not None:
        break
```

```python
filter_kernel = np.array([[1, 1], [1, -1]])
```

```python
feature_map = np.zeros((2, 2), dtype=int)
for r in range(2):
    for c in range(2):
        feature_map[r, c] = np.sum(patch[r:r+2, c:c+2] * filter_kernel)
```

```python
print("Image Patch:")
print(patch)
print("\nFilter:")
print(filter_kernel)
print("\nTop-left Convolution Output:", feature_map[0, 0])
print("\nFeature Map:")
print(feature_map)
```

```python
sample_patch = np.array([[252, 252, 220], [128, 58, 22], [0, 0, 0]])
sample_filter = np.array([[1, 1], [1, -1]])
sample_feature_map = np.zeros((2, 2), dtype=int)
for r in range(2):
    for c in range(2):
        sample_feature_map[r, c] = np.sum(sample_patch[r:r+2, c:c+2] * sample_filter)
```

```python
print("Sample Image Patch:")
print(sample_patch)
print("\nSample Filter:")
print(sample_filter)
print("\nSample Top-left Convolution Output:", sample_feature_map[0, 0])
print("\nSample Feature Map:")
print(sample_feature_map)
```
