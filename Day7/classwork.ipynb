```python
kernel = np.array([
    [1, 1, 1],
    [1, 1, 1],
    [1, 1, 1]
], dtype=np.float32)
```

```python
def convolution(image, kernel, stride=1, padding=0):
    image = np.pad(image, pad_width=padding, mode='constant', constant_values=0)

    img_height = image.shape[0]
    img_width = image.shape[1]

    output_height = (img_height - kernel.shape[0] + 2 * padding) // stride + 1
    output_width = (img_width - kernel.shape[1] + 2 * padding) // stride + 1

    output = np.zeros((output_height, output_width), dtype=np.float32)

    for y in range(output_height):
        for x in range(output_width):
            patch = image[y * stride: y * stride + kernel.shape[0], x * stride: x * stride + kernel.shape[1]]
            output[y, x] = np.sum(patch * kernel)
    return output
```

```python
output = convolution(gray_image, kernel, stride=1, padding=0)
plt.imshow(output, cmap='gray')
```
