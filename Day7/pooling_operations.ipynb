```python
import numpy as np
from tensorflow.keras.datasets import fashion_mnist
```

```python
(fashion_x, _), _ = fashion_mnist.load_data()
print(fashion_x.shape)
```

```python
def max_pooling2d(image, pool_size=(2, 2), stride=2):
    img_h, img_w = image.shape
    ph, pw = pool_size
    out_h = (img_h - ph) // stride + 1
    out_w = (img_w - pw) // stride + 1
    output = np.zeros((out_h, out_w), dtype=image.dtype)
    for r in range(out_h):
        for c in range(out_w):
            r_start = r * stride
            c_start = c * stride
            output[r, c] = np.max(image[r_start:r_start+ph, c_start:c_start+pw])
    return output

def avg_pooling2d(image, pool_size=(2, 2), stride=2):
    img_h, img_w = image.shape
    ph, pw = pool_size
    out_h = (img_h - ph) // stride + 1
    out_w = (img_w - pw) // stride + 1
    output = np.zeros((out_h, out_w), dtype=np.float32)
    for r in range(out_h):
        for c in range(out_w):
            r_start = r * stride
            c_start = c * stride
            output[r, c] = np.mean(image[r_start:r_start+ph, c_start:c_start+pw])
    return output
```

```python
input_patch = np.array([
    [0, 0, 0, 193],
    [3, 0, 12, 219],
    [6, 0, 99, 244],
    [0, 0, 55, 236]
])

max_pooled = max_pooling2d(input_patch)
avg_pooled = avg_pooling2d(input_patch)

print("Input Patch:")
print(input_patch)
print("\nMax Pooling Output:")
print(max_pooled)
print("Shape:", max_pooled.shape)
print("\nAverage Pooling Output:")
print(avg_pooled)
print("Shape:", avg_pooled.shape)
```

```python
single_image = fashion_x[0]
extracted_patch = single_image[10:14, 10:14]

fashion_max_pooled = max_pooling2d(extracted_patch)
fashion_avg_pooled = avg_pooling2d(extracted_patch)

print("Extracted Fashion-MNIST Patch:")
print(extracted_patch)
print("\nMax Pooled Output:")
print(fashion_max_pooled)
print("Shape:", fashion_max_pooled.shape)
print("\nAverage Pooled Output:")
print(fashion_avg_pooled)
print("Shape:", fashion_avg_pooled.shape)
```
