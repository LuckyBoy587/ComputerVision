```python
inputs = [2.0, -1.0]
weights = [0.4, 0.6]
bias = -0.2
```

```python
Z = bias + (inputs[0] * weights[0] + inputs[1] * weights[1])
Z
```

```python
def relu(val):
    return max(0, val)
```

```python
relu(Z)
```

```python
X = [-2.0, 1.0]
```

```python
Z = bias + (X[0] * weights[0] + X[1] * weights[1])
Z
```

```python
relu(Z)
```
