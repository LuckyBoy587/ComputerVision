```
import numpy as np
```

```
z = -1
```

```
def wrong_sigmoid(z):
    return 1 / (1 + np.exp(z))

def correct_sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```
wrong_out = wrong_sigmoid(z)
correct_out = correct_sigmoid(z)
print(f"Wrong Sigmoid Output: {wrong_out:.3f}")
print(f"Correct Sigmoid Output: {correct_out:.3f}")
```
