```
import numpy as np
```

```
x = np.array([1.5, -2.0])
w1 = np.array([0.3, -0.8])
b1 = 0.2
w2 = 0.7
b2 = -0.1
```

```
def relu(z):
    return np.maximum(0, z)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def forward_propagation(x, w1, b1, w2, b2):
    z1 = np.dot(x, w1) + b1
    a1 = relu(z1)
    z2 = a1 * w2 + b2
    a2 = sigmoid(z2)
    return z1, a1, z2, a2
```

```
z1, a1, z2, a2 = forward_propagation(x, w1, b1, w2, b2)
print(f"Z1 = {z1:.2f}")
print(f"A1 (ReLU) = {a1:.2f}")
print(f"Z2 ≈ {z2:.3f}")
print(f"Final Output (Sigmoid) ≈ {a2:.4f}")
```
