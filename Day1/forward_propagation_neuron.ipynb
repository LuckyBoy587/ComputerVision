```python
import numpy as np
```

```python
X = np.array([1.0, 2.0])
W = np.array([0.5, -0.6])
b = 0.1
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def relu(z):
    return np.maximum(0, z)

def forward_neuron(x, w, bias):
    z = np.dot(w, x) + bias
    return z, sigmoid(z), relu(z)
```

```python
z, sigmoid_out, relu_out = forward_neuron(X, W, b)
print(f"Z = {z:.1f}")
print(f"Sigmoid Output ≈ {sigmoid_out:.3f}")
print(f"ReLU Output = {int(relu_out)}")
```
