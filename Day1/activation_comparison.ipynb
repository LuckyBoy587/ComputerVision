```
import numpy as np
```

```
x = np.array([0.5, 1.5])
w = np.array([-0.4, 0.9])
b = 0.2
```

```
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def relu(z):
    return np.maximum(0, z)

def tanh(z):
    return np.tanh(z)

def compute_neuron(x, w, b):
    z = np.dot(x, w) + b
    return z, sigmoid(z), relu(z), tanh(z)
```

```
z, sig_out, relu_out, tanh_out = compute_neuron(x, w, b)
print(f"Z = {z:.2f}")
print(f"Sigmoid Output ≈ {sig_out:.3f}")
print(f"ReLU Output = {relu_out:.2f}")
print(f"Tanh Output ≈ {tanh_out:.3f}")
```
