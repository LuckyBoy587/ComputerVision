```python
inputs = [1.0, 2.0]
weights = [0.5, -0.6]
bias = 0.1
```

```python
output_neuron = sum(inputs[i] * weights[i] for i in range(len(inputs))) + bias
output_neuron
```

```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
sigmoid(output_neuron)
```

```python
def relu(x):
    return max(0, x)
```

```python
relu(output_neuron)
```

```python
import numpy as np

def forward_propagation(inputs, weights, bias):
    wx = np.dot(inputs, weights)
    z = wx + bias

    sigmoid_output = sigmoid(z)
    relu_output = relu(z)
    return z, sigmoid_output, relu_output
```

```python
forward_propagation(inputs, weights, bias)
```
