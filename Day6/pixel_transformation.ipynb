```python
from tensorflow.keras.datasets import fashion_mnist
(fashion_x, _), _ = fashion_mnist.load_data()
print(fashion_x.shape)
```

```python
import cv2
import numpy as np

sample_images = fashion_x[:5]

bright_images = [np.clip(img.astype(np.uint16) + 50, 0, 255).astype(np.uint8) for img in sample_images]
blurred_images = [cv2.GaussianBlur(img, (5, 5), 0) for img in bright_images]
edge_images = [cv2.Canny(img, 50, 150) for img in blurred_images]
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(4, 5, figsize=(10, 8))

for i in range(5):
    axes[0, i].imshow(sample_images[i], cmap='gray')
    axes[0, i].set_title("Original", fontsize=9)
    axes[0, i].axis('off')

for i in range(5):
    axes[1, i].imshow(bright_images[i], cmap='gray')
    axes[1, i].set_title("Brightened", fontsize=9)
    axes[1, i].axis('off')

for i in range(5):
    axes[2, i].imshow(blurred_images[i], cmap='gray')
    axes[2, i].set_title("Blurred", fontsize=9)
    axes[2, i].axis('off')

for i in range(5):
    axes[3, i].imshow(edge_images[i], cmap='gray')
    axes[3, i].set_title("Edges", fontsize=9)
    axes[3, i].axis('off')

plt.tight_layout()
plt.show()
```
