```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.datasets import mnist, cifar10
```

```python
(mnist_x, _), _ = mnist.load_data()
(cifar_x, _), _ = cifar10.load_data()

mnist_imgs = mnist_x[:5]
cifar_imgs = cifar_x[:5]
```

```python
for i, img in enumerate(cifar_imgs):
    r, g, b = img[16, 16]
    gray = (int(r) + int(g) + int(b)) / 3
    print(f"Image {i+1}: R={r}, G={g}, B={b} \u2192 Gray={round(gray, 2)}")
```

```python
fig, axes = plt.subplots(3, 5, figsize=(10, 6))

for i in range(5):
    axes[0, i].imshow(mnist_imgs[i], cmap='gray')
    axes[0, i].set_title("MNIST")
    axes[0, i].axis('off')

for i in range(5):
    axes[1, i].imshow(cifar_imgs[i])
    axes[1, i].set_title("CIFAR-10")
    axes[1, i].axis('off')

for i in range(5):
    gray_img = cv2.cvtColor(cifar_imgs[i], cv2.COLOR_RGB2GRAY)
    axes[2, i].imshow(gray_img, cmap='gray')
    axes[2, i].set_title("CIFAR-10 Gray")
    axes[2, i].axis('off')

plt.tight_layout()
plt.show()
```
