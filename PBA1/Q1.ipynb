```python
import numpy as np
import matplotlib.pyplot as plt

dl_dw1 = lambda val: 2 * val
dl_dw2 = lambda val: 20 * val
```

```python
lr = 0.05
w1, w2 = 8.0, 4.0
exp1_w1 = [w1]
exp1_w2 = [w2]

for t in range(50):
    w1 = w1 - lr * dl_dw1(w1)
    w2 = w2 - lr * dl_dw2(w2)
    exp1_w1.append(w1)
    exp1_w2.append(w2)
```

```python
lr = 0.09
w1, w2 = 8.0, 4.0
exp2_w1 = [w1]
exp2_w2 = [w2]

for t in range(50):
    w1 = w1 - lr * dl_dw1(w1)
    w2 = w2 - lr * dl_dw2(w2)
    exp2_w1.append(w1)
    exp2_w2.append(w2)
```

```python
w1, w2 = 8.0, 4.0
exp3_w1 = [w1]
exp3_w2 = [w2]

for t in range(50):
    lr = 0.05 / (1 + 0.1 * t)
    w1 = w1 - lr * dl_dw1(w1)
    w2 = w2 - lr * dl_dw2(w2)
    exp3_w1.append(w1)
    exp3_w2.append(w2)
```

```python
w1_vals = np.linspace(-10, 10, 400)
w2_vals = np.linspace(-5, 5, 400)
W1, W2 = np.meshgrid(w1_vals, w2_vals)
L = W1**2 + 10 * W2**2

plt.figure(figsize=(10, 8))
contours = plt.contour(W1, W2, L, levels=30, cmap='viridis')
plt.clabel(contours, inline=True, fontsize=8)

plt.plot(exp1_w1, exp1_w2, 'r-o', label='Experiment A (alpha = 0.05)', markersize=3)
plt.plot(exp2_w1, exp2_w2, 'b-x', label='Experiment B (alpha = 0.09)', markersize=3)
plt.plot(exp3_w1, exp3_w2, 'g-d', label='Experiment C (decay)', markersize=3)

plt.plot(8.0, 4.0, 'ko', markersize=8, label='Start (8.0, 4.0)')

plt.plot(exp1_w1[-1], exp1_w2[-1], 'ro', markersize=8, markeredgecolor='black', label=f'End A ({exp1_w1[-1]:.2f}, {exp1_w2[-1]:.2f})')
plt.plot(exp2_w1[-1], exp2_w2[-1], 'bx', markersize=8, markeredgecolor='black', label=f'End B ({exp2_w1[-1]:.2f}, {exp2_w2[-1]:.2f})')
plt.plot(exp3_w1[-1], exp3_w2[-1], 'gd', markersize=8, markeredgecolor='black', label=f'End C ({exp3_w1[-1]:.2f}, {exp3_w2[-1]:.2f})')

plt.xlabel('w1')
plt.ylabel('w2')
plt.title('Gradient Descent Trajectories on 2D Loss Surface')
plt.legend()
plt.grid(True)
plt.show()
```

### Task 4: Analysis

#### 1. Which trajectory converges smoothest?
Experiment C (using learning rate decay) converges the smoothest. Although Experiment A ($\alpha = 0.05$) converges, it exhibits oscillations in the beginning along the $w_2$ direction. Experiment C reduces the learning rate step-by-step, allowing rapid progress initially and then suppressing oscillations as it approaches the minimum, leading to a smooth path to the optimum.

#### 2. What happens to $w_2$ under Experiment B? Why?
Under Experiment B ($\alpha = 0.09$), the trajectory for $w_2$ oscillates wildly back and forth across the valley.
- The gradient along $w_2$ is $\frac{\partial L}{\partial w_2} = 20 w_2$.
- The update equation is $w_2^{(t+1)} = w_2^{(t)} - \alpha (20 w_2^{(t)}) = w_2^{(t)}(1 - 20\alpha)$.
- For $\alpha = 0.09$, the scaling factor is $1 - 20(0.09) = 1 - 1.8 = -0.8$.
- Since the factor is negative and has an absolute value close to $1.0$ ($|-0.8| < 1.0$), $w_2$ alternates signs and oscillates heavily across the valley, but still slowly decays toward $0.0$. If $\alpha \ge 0.1$, the updates would diverge or oscillate indefinitely, leading to complete instability along the $w_2$ axis.

#### 3. How does learning rate decay help on ill-conditioned surfaces?
An ill-conditioned surface (like this elongated bowl) has a much larger gradient along one axis ($w_2$ with weight 10) than the other ($w_1$ with weight 1). A constant high learning rate suited for $w_1$ causes instability and wild oscillations along $w_2$. A constant low learning rate avoids oscillations along $w_2$ but makes progress along $w_1$ extremely slow.
Learning rate decay solves this by dynamically decreasing the learning rate over time:
- In early steps, a larger learning rate allows fast progress along the flatter $w_1$ direction.
- As the learning rate decays, it becomes small enough to prevent/dampen oscillations along the steep $w_2$ direction, allowing the optimizer to safely settle into the minimum.
