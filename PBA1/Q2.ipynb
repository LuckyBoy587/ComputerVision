```python
import numpy as np
```

```python
X = np.array([2.0, 3.0])
W = np.array([0.5, -0.4])
b = 1.0
```

```python
Z = W * X + b
Z
```

```python
def relu(arr):
    return np.maximum(0, arr)
```

```python
A = relu(Z)
A
```
