```python
true_val = 1
pred_val = 0.8
```

```python
def mse(y_true, y_pred):
    return (y_true - y_pred) ** 2
```

```python
import math

def bce(y_true, y_pred):
    return -(y_true * math.log(y_pred) + (1 - y_true) * math.log(1 - y_pred))
```

```python
print(f"MSE: {mse(true_val, pred_val):.2f}")
print(f"BCE: {bce(true_val, pred_val):.4f}")
```
