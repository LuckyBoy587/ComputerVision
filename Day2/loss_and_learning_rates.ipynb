```python
import numpy as np
```

```python
y = 1.0
y_hat = 0.6
w = 0.8
dw = 0.5
eta_1 = 0.1
eta_2 = 0.5
```

```python
def compute_mse(y, y_hat):
    return (y - y_hat) ** 2


def compute_bce(y, y_hat):
    return - (y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))


def update_weight(w, eta, dw):
    return w - eta * dw
```

```python
mse = compute_mse(y, y_hat)
bce = compute_bce(y, y_hat)
w_updated_1 = update_weight(w, eta_1, dw)
w_updated_2 = update_weight(w, eta_2, dw)

print(f"MSE = {mse:.2f}")
print(f"Binary Cross Entropy ≈ {bce:.4f}")
print(f"Updated weight (η = {eta_1}) = {w_updated_1:.2f}")
print(f"Updated weight (η = {eta_2}) = {w_updated_2:.2f}")
```
