```
y = 0.0
y_hat = 0.3
x = 1.5
w = 0.6
learning_rate = 0.2
```

```
def compute_dL_dyhat(y, y_hat):
    return (y_hat - y) / (y_hat * (1 - y_hat))

def compute_dyhat_dw(x):
    return x

def update_weight(w, learning_rate, dL_dw):
    return w - learning_rate * dL_dw
```

```
dL_dyhat = compute_dL_dyhat(y, y_hat)
dyhat_dw = compute_dyhat_dw(x)
dL_dw = dL_dyhat * dyhat_dw
w_updated = update_weight(w, learning_rate, dL_dw)

print(f"dL/dŷ ≈ {dL_dyhat:.4f}")
print(f"dŷ/dw = {dyhat_dw:.1f}")
print(f"dL/dw ≈ {dL_dw:.4f}")
print(f"Updated weight ≈ {w_updated:.4f}")
```
