```python
from tensorflow.keras import layers, models
```

```python
model = models.Sequential([
    layers.Input(shape=(28, 28, 1)),
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
print("Number of Convolution Layers:", 1)
print("Number of Pooling Layers:", 1)
print("Number of Dense Layers:", 2)
```

```python
import cv2
import numpy as np

img = cv2.imread('/content/ComputerVision/data/img.png', cv2.IMREAD_GRAYSCALE)
img_resized = cv2.resize(img, (28, 28))
img_normalized = img_resized / 255.0
x_train = np.expand_dims(img_normalized, axis=(0, -1))
y_train = np.array([1.0])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=5)
```
