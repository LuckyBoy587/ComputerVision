```python
import cv2
image = cv2.imread('/content/ComputerVision/data/sample.jpg')
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
```

```python
image_800x600 = cv2.resize(image, (800, 600))
```

```python
print(f"Original Image Resolution: {image_800x600.shape[0]}x{image_800x600.shape[1]}")
```

```python
image_224x224 = cv2.resize(image, (224, 224))
```

```python
print(f"Resized Image Resolution: {image_224x224.shape[0]}x{image_224x224.shape[1]}")
```

```python
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image_800x600, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Original Image")

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(image_224x224, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Resized Image")

plt.show()
```

```python
import numpy as np
print("Total pixels in original image:", np.prod(image_800x600.shape[:2]))
print("Total pixels in resized image:", np.prod(image_224x224.shape[:2]))
```
