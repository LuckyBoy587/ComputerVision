### Task 1: Theory

#### Why ordinary neural networks are not suitable for sequence data:
1. **Fixed Input and Output Dimensions**: Standard feedforward neural networks require a fixed-size input vector and produce a fixed-size output vector. Sequence data, such as sentences or time series, typically has variable lengths.
2. **Lack of Temporal Memory**: Ordinary neural networks process each input independently of previous inputs. They do not maintain a state or memory to store information from past elements in the sequence, which is crucial for understanding context.
3. **No Parameter Sharing Across Time**: Standard networks do not share parameters across different positions in a sequence. A feature learned at one time step cannot easily be applied to another time step.

#### How LSTM overcomes the limitations of RNN:
1. **Vanishing/Exploding Gradients**: Standard Recurrent Neural Networks (RNNs) suffer from vanishing or exploding gradients when trained on long sequences, making it difficult to learn long-term dependencies.
2. **Gating Mechanisms**: Long Short-Term Memory (LSTM) networks introduce a cell state regulated by three gates:
   - **Forget Gate**: Decides what information to discard from the cell state.
   - **Input Gate**: Decides what new information to store in the cell state.
   - **Output Gate**: Decides what the next hidden state should be based on the cell state.
3. **Long-Term Dependency Preservation**: These gates allow gradients to flow backwards through time without vanishing quickly, enabling LSTMs to retain memory over long sequences.

```python
import numpy as np
import tensorflow as tf
```

```python
x = np.random.rand(100, 5, 1)
y = np.random.rand(100, 1)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=(5, 1)),
    tf.keras.layers.LSTM(32),
    tf.keras.layers.Dense(1)
])

model.compile(
    optimizer='adam',
    loss='mse'
)

model.summary()
```

```python
history = model.fit(x, y, epochs=10, batch_size=8)
```
