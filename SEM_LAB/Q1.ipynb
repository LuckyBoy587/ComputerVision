```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../data/Q1.png")
plt.figure(figsize=(10, 10))
plt.title("Original Image")
plt.imshow(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10, 10))
plt.title("Gray Image")
plt.imshow(gray_image, cmap="gray")
```

```python
resize_image = cv2.resize(gray_image, (224, 224))
plt.figure(figsize=(10, 10))
plt.title("Resize image")
plt.imshow(resize_image, cmap="gray")
```

```python
canny_edge_image =  cv2.Canny(resize_image, 100, 200)
plt.figure(figsize=(10, 10))
plt.title("Canny Image")
plt.imshow(canny_edge_image, cmap="gray")
```
