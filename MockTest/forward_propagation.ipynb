```python
import numpy as np
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def relu(z):
    return np.maximum(0, z)

def tanh(z):
    return np.tanh(z)
```

```python
def forward_pass(X, layers):
    cache = []
    A = X
    for layer in layers:
        W = layer["W"]
        b = layer["b"]
        activation_name = layer["activation"]
        Z = np.dot(W, A) + b
        if activation_name == "sigmoid":
            A = sigmoid(Z)
        elif activation_name == "relu":
            A = relu(Z)
        elif activation_name == "tanh":
            A = tanh(Z)
        cache.append({"Z": Z, "A": A, "activation": activation_name})
    return A, cache
```

```python
X = np.array([1.0, -0.5, 2.0])

W1 = np.array([0.5, 0.2, 0.4])
b1 = np.array([0.15])

W2 = np.array([0.5])
b2 = np.array([0.105])

W3 = np.array([0.2])
b3 = np.array([0.08946])

layers_relu = [
    {"W": W1, "b": b1, "activation": "relu"},
    {"W": W2, "b": b2, "activation": "relu"},
    {"W": W3, "b": b3, "activation": "sigmoid"}
]

layers_tanh = [
    {"W": W1, "b": b1, "activation": "relu"},
    {"W": W2, "b": b2, "activation": "tanh"},
    {"W": W3, "b": b3, "activation": "sigmoid"}
]
```

```python
output_relu, cache_relu = forward_pass(X, layers_relu)
for i, step in enumerate(cache_relu):
    print(f"Layer {i+1} | Z = {np.round(step['Z'][0], 3)} | A ({step['activation'].capitalize()}) = {np.round(step['A'][0], 3)}")
```

```python
output_tanh, cache_tanh = forward_pass(X, layers_tanh)
for i, step in enumerate(cache_tanh):
    print(f"Layer {i+1} | Z = {np.round(step['Z'][0], 3)} | A ({step['activation'].capitalize()}) = {np.round(step['A'][0], 3)}")
```
