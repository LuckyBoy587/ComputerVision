```python
import numpy as np
```

```python
input_features = 3
hidden_neurons = 4
output_neurons = 1
dropout_rate = 0.5
X = np.array([2, 6, 10, 14])
```

```python
def compute_weights(in_feat, hid_neu, out_neu):
    w_input_to_hidden = in_feat * hid_neu
    w_hidden_to_output = hid_neu * out_neu
    return w_input_to_hidden + w_hidden_to_output

def compute_active_neurons(hid_neu, drop_rate):
    return hid_neu * (1.0 - drop_rate)

def compute_mean(data):
    return np.mean(data)

def compute_variance(data):
    return np.var(data)
```

```python
total_weights = compute_weights(input_features, hidden_neurons, output_neurons)
active_neurons = compute_active_neurons(hidden_neurons, dropout_rate)
mean_val = compute_mean(X)
var_val = compute_variance(X)

print(f"Total weights = {total_weights}")
print(f"Active neurons after dropout = {active_neurons}")
print(f"Mean = {mean_val}")
print(f"Variance = {var_val}")
```
