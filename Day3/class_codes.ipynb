```python
x = np.linspace(-10, 10, 200)
```

```python
def sigmoid(val):
    return 1 / (1 + np.exp(-val))
```

```python
sig_x = sigmoid(x)
```

```python
def relu(val):
    return np.maximum(0, val)
```

```python
relu_x = relu(x)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, sig_x, label="sigmoid")
plt.plot(x, relu_x, label="relu")
plt.legend()
plt.grid(True)
```
