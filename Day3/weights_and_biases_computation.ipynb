```
input_features = 4
hidden_1 = 5
hidden_2 = 3
output_neurons = 1
```

```
def compute_dense_params(inputs, outputs):
    weights = inputs * outputs
    biases = outputs
    return weights, biases
```

```
w1, b1 = compute_dense_params(input_features, hidden_1)
w2, b2 = compute_dense_params(hidden_1, hidden_2)
w3, b3 = compute_dense_params(hidden_2, output_neurons)

print(f"Input → Hidden Layer 1: Weights = {w1}, Biases = {b1}")
print(f"Hidden Layer 1 → Hidden Layer 2: Weights = {w2}, Biases = {b2}")
print(f"Hidden Layer 2 → Output Layer: Weights = {w3}, Biases = {b3}")
```
