```python
import numpy as np
```

```python
total_neurons = 8
dropout_rate = 0.25
X = [1, 3, 5, 7, 9]
```

```python
def compute_active_neurons(total, rate):
    return total * (1.0 - rate)

def compute_mean_and_variance(data):
    mean = np.mean(data)
    variance = np.var(data)
    return mean, variance
```

```python
active_neurons = compute_active_neurons(total_neurons, dropout_rate)
mean, variance = compute_mean_and_variance(X)
print(f"Active neurons after dropout = {active_neurons}")
print(f"Mean = {mean}")
print(f"Variance = {variance}")
```
